## Score Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
model_df = pd.read_csv('model_perf.csv')
```

```python
log_reg_res = model_df[model_df['model_type'] == 'Logistic Regression']
fig,axes = plt.subplots(figsize=(15,10))
plt.plot(log_reg_res[log_reg_res['C']==0.01]['no_of_features'],log_reg_res[log_reg_res['C']==0.01]['AUC Score'])
plt.plot(log_reg_res[log_reg_res['C']==0.1]['no_of_features'],log_reg_res[log_reg_res['C']==0.1]['AUC Score'])
plt.plot(log_reg_res[log_reg_res['C']==1]['no_of_features'],log_reg_res[log_reg_res['C']==1]['AUC Score'])
plt.plot(log_reg_res[log_reg_res['C']==10]['no_of_features'],log_reg_res[log_reg_res['C']==10]['AUC Score'])
plt.plot(log_reg_res[log_reg_res['C']==100]['no_of_features'],log_reg_res[log_reg_res['C']==100]['AUC Score'])
plt.legend(['0.01','0.1','1','10','100'])
plt.title('Logistic Regression Hyper Parameter Tuning')
plt.xlabel('No of Features')
plt.ylabel('AUC Score')

```
